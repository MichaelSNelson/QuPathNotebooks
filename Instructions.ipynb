### Project 1 setup
1) Getting started
a.	Create empty folder 
b.	Drag into QuPath. Yes, create project.
c.	Drag in lung images


2)	Open Img1<br>



a.	Change to autoestimate<br>
b.	Apply<br>
<br>
It will look too bright!<br>




3) Visualization Settings

a.	Open Brightness/Contrast <br>


b.	Click Show Grayscale<br>
 
c.	Make sure Nucleus_Hoescht is selected<br>
d.	Zoom in. Adjust Minimum and Maximum until the non-tissue background is gone and nuclei are just saturated enough. <br>
e.	Repeat for all channels<br>
f.	For blank channels (Argo874 and Argo890), I like to move both the min and max to a very high value so that they never accidentally get turned on.<br>
g.	Turn on everything except the blank channels. Now it looks better, but still confusing.<br>
<br>
h.	Save Display Settings.<br>
<br>


4)	Create different sets of visualization settings- Cancer, T cells, Myeloid, etc. <br>
a.	Click on colored square to change channel color. <br>
b.	Play with gamma to see how it changes, but be careful!<br>
c.	Keeps saving them with descriptive names<br>


5)	CREATE ROI FOR EXPORT<br>
a.	Objects > Annotations > Specify Annotation<br>
b.	Choose um<br>
c.	Give it a name<br>
<br>
e.	Click Add annotation<br>
f.	Move it to somewhere interesting.<br>
g.	With box selected, File > Export images > Rendered RGB (with overlays)<br>
h.	Change to Tif, choose appropriate downsample<br>
i.	Navigate to project folder<br>
j.	Create exports folder<br>
k.	Name by region name_dimension_ display settings<br>
l.	Change display settings to a different set<br>
m.	Export again with new colors<br>
n.	Look, side-by-side matched panels!<br>


6)	One time set up for scripts<br>
a.	Settings - the icon should be in the upper right and look like this! If your monitor resolution is too low, you may need to click on the >> at the top right of QuPath to show this button, and others.<br>
<br>
b.	Search Scripts<br>
c.	Set Script Directory<br>
 <br>

d.	Put ChannelNamesAndColors.groovy into that folder<br>


7)	Export with channel names and colors<br>
a.	Automate > Shared Scripts > ChannelNames<br>
b.	In Automation window, Run > Run<br>
c.	Yay for a useful window!<br>
d.	File > Export Snapshot > Main Window Screenshot<br>
