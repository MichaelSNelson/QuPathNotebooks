### 2. Use manual annotation tools to define an area
We're going to try to mark the cancer area manually to practice using the Annotation tools. We'll use these skills to learn a more automated way later. Practice with these tool to mark the S100+ region. 
1. Brush:  <img src="https://qupath.readthedocs.io/en/stable/_images/BRUSH_TOOL.png" width = "50">
2. Polygon:  <img src="https://qupath.readthedocs.io/en/stable/_images/POLYGON_TOOL.png" width = "50">
3. Wand: <img src="https://qupath.readthedocs.io/en/stable/_images/WAND_TOOL.png" width = "50">

Things to note:
  - Hit <kbd>Alt</kbd> to turn the brush or wand into an eraser!
  - You can mix and match the tools whenever convenient. I like to add with brush and then clean up mistakes with wand.
  - While drawing, you can hit <kbd>Ctrl</kbd> + <kbd>Shift</kbd> to stop an annotation at an existing boundary. 
  
For more information about annotations, see here: https://qupath.readthedocs.io/en/stable/docs/starting/annotating.html#brush-tool

#### 3. Combine the areas to measure tumor burden
1. Your Annotations tab will look something like this: 

2. Select everything except the Region* annotation that outlines the entire tissue. `Right click > Edit multiple > Merge selected`
3. Select the new annotation that defines the entire cancer region. Then, in the Class list on the right, select Tumor.
<img src="TumorClass.PNG">

4. Click `Set Selected` to assign that annotation the class "Tumor".
5. Click on the tumor annotation, at the bottom of the Annotations tab, look at the total S100 Area. In your favorite calculator program, divide that by the total tissue area to quantify the cancer burden.  
  - My results are $168700 / 476700 = 0.35%$. I.E., 35% of the tissue area is positive for S100.
